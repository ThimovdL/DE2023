# Superstore dataset

## Setting up

```python
#Load the libraries
import pandas as pd
import chardet
```

```python
#import the data
file_path = 'Superstore.csv'

# Detect the file encoding
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

# Use the detected encoding
superstore = pd.read_csv(file_path, encoding=result['encoding'])
```

```python
#check the data
superstore.head(10)
```
